#### Task 9 
Analyze songs by artist and views-youtube, stream-spotify.

```python
df_Top10 = df.groupby(['Artist'])[['Views','Stream']].sum()
df_Top10
```

```python
df_top10_youtube = df_Top10.sort_values('Views', ascending = False)
df_top10_youtube
```

```python
df_top10_spotify = df_Top10.sort_values('Stream', ascending = False)
df_top10_spotify
```

#### Task 10
Create bar charts for Top 10 Artists on Youtube and Top 10 Artists on Spotify in 1 row of display.

```python
df_youtube_top10 = df_top10_youtube[["Views"]].head(10)
df_youtube_top10
```

```python
df_spotify_top10 = df_top10_spotify[["Stream"]].head(10)
df_spotify_top10
```

```python
fig,(youtube,spotify)  = plt.subplots(ncols = 2, figsize = (20,8))

df_youtube_top10.plot.bar(y= 'Views', rot = 80, title = 'Top 10 Artist on Youtube', ax = youtube)
youtube.legend([])

df_spotify_top10.plot.bar(y= 'Stream', rot = 80, title = 'Top 10 Artist on Spotify', ax = spotify)
spotify.legend([])

plt.tight_layout()
plt.show()
```

#### Task 15
Display a bar chart of the Top 10 Songs by Stream with their Energy, Danceability, and Acousticness and make some remarks about it.

```python
df_stream = df.groupby(['Track'])[['Stream','Energy','Danceability','Acousticness']].sum().sort_values('Stream', ascending = False).head(10)
df_stream
```

```python
df_stream[['Stream']].plot.bar(y = 'Stream', rot = 80, title = 'Top 10 Most Streamed on Spotify', figsize = (20,8))
plt.legend([]) 
# The stream value is too high to be together with the energy danceability and acousticness for that reason we seperate them
bar_EDA = df_stream[['Energy','Danceability','Acousticness']].plot.bar(stacked = False, figsize = (20,8), rot = 80)
bar_EDA.set_title('Song Character')
plt.show()
```

### Remark:
if we were to cut the graph at 0.5 we can see tahat the most consistent bar is those that has more danceability aspect to thier music and the most suprising this is that there are a lot of songs that doesn't have that acousticness but stil are able to make it to be one of the most streamed songs. As for energy they mostly vary on the theme that the artist has intended but if we were to cut the graph at 0.5 again we would see that at the very least a song must have energy.
