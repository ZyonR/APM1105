```python
#Number 1 SA1
import requests
import random
import os
 
url = 'https://raw.githubusercontent.com/AllenDowney/ThinkPython2/master/code/words.txt'
request = requests.get(url)

if request.status_code == 200:
    with open('words.txt', 'wb') as f:
        f.write(request.content)
        print('Dataset saved to file.')
else:
    print('Failed to download dataset.')

words = list(request.text.split())
#Letter A
print(f'Total number of words: {len(words)}')

#Letter B
words_dupe = []
for word in words:
    words_dupe.append(len(word))
words_dupe.sort()
print("\n")
for word in words:
    if len(word) == words_dupe[-1]:
        print(f"Longest word: {word}")

#Letter C
n_letter = 0
n_input = input("\n\nEnter a letter: ")
n_input = n_input.lower()
for word in words:
    if word[0]== n_input:
        n_letter+=1
    else:
        continue
print(f"The number of words that starts with {n_input} is {n_letter}")

#Letter D
reverse_words = []
for word in words:
    reverse_words.append(word[::-1])
palindromes = []
counter = 0
while counter!= len(words):
    if reverse_words[counter]==words[counter]:
        palindromes.append(words[counter])
        counter+=1
    else:
        counter+=1
print(palindromes)

#Letter E
frequency_counter = 0
input_letter_frequency = input("\n\nInput a letter: ")
input_letter_frequency = input_letter_frequency.lower()
for word in words:
    if  word.find(input_letter_frequency) < 0 :
        continue
    else:
        frequency_counter += word.find(input_letter_frequency)
print(f"\n\nThe frequency of the letter {input_letter_frequency} is {frequency_counter}")

#Letter F
input_substring = input("Input substring: ")
print(f"\n\nWords that contain the substring {input_substring} are: ")
for word in words:
    subString_count = word.find(input_substring)
    if subString_count > 0:
        print(word,end=", ")

#Letter G
def chooseRandom ():
    random_word = random.choice(words)
    return random_word
guess_me = []
def underscorer():
    random_word = chooseRandom()
    for letter in random_word:
        guess_me.append("_")
    return random_word
def game(randomWord):
    lifeAmount = 7
    itEquates = 0
    game_over = False
    while lifeAmount != -1 and not game_over:
        if lifeAmount == 0:
            guess_me.clear()
            os.system('cls')
            print(f"\nSorry no lives left ðŸ˜ž")
            print(f"Word was: {randomWord}\n")
            break
        else:
            print("\n")
            for line in guess_me:
                print(line, end=" ")
            if lifeAmount > 0:
                print(f"Remaining Life: {lifeAmount}")
                guess_input = input("Input Letter: ")
                guess_input = guess_input.lower().strip()
                if guess_input in guess_me:
                    continue
                else:
                    if randomWord.find(guess_input) >= 0:
                        enumerated = enumerate(randomWord)
                        for index, letter in enumerated:
                            if letter == guess_input:
                                guess_me[index] = guess_input
                                itEquates+=1
                                if itEquates == len(randomWord):
                                    guess_me.clear()
                                    os.system('cls')
                                    print("Congratulations you got it correct!!!")
                                    print(f"Word was indeed {randomWord}")
                                    game_over = True
                                    break
                    else:
                        lifeAmount -=1
            else:
                guess_me.clear()
                os.system('cls')
                print(f"\nSorry no lives left ðŸ˜ž")
                print(f"Word was: {randomWord}\n")
                break
def main ():
    print("\n\n\nWelcome User, to Fill-in-the-blanks!!!")
    random_word = underscorer()
    game(random_word)
    ask = input("\nWould you like to try again?(y|n): ")
    ask = ask.lower()
    match ask: 
        case "y"|"yes":
            os.system('cls')
            main()
        case "n"|"no":
            print("Thank You for playing!!!")
main()
```
